## Capstone project. FINAL REPORT

### 1. BUSINESS PROBLEM

This is a business problem for all late night coffee lovers. It has been discovered, that if one wants to find a high-end unique coffee place in later hours - not necessarily an easy task. The customer is a person who likes a good cup of coffee in the afterhours in a big city in EU, maybe a travelling business man, maybe a coffee enthusiast. Additionally, results might reveal bussiness possibilities for somebody who is looking to open a Coffee Shop with longer work hours.

#### The following will summarize criteria (business problem requirements) for the data we will be interested in:

* Coffee shops need to work late hours (to yield sufficient data set, closing time should be no later than 7 PM)
* Coffee shops need to have rating 8 or more to be worth the while
* Coffee shops need to be placed in a Capital in the EU, within 2km radius of the center

### 3. METHODOLOGY

Used mathodology: exploratory data analysis

First, simply looking into data to determine:
* Top 3 capitals with most high-rating late working Coffee Shops
* Top 10 later working Coffee Shops with highest ratings

Secondly, kMeans algorithm was used to cluster the Coffee Shops into 3 clusters to find best distance_from_center/opening_hours/rating ratio for the Coffee Shops
